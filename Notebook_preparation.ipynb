### Domaine 3 : création indices (pour calculer efficacité gouvernementale)

Taux de mortalité

```python
plt.figure(figsize=(7,7))
sns.histplot(x=mortality_rate['Mortality rate attributed to exposure to unsafe WASH services'], bins=1000)
plt.xlim(0,10)
plt.show()
mortality_rate.describe()
```

```python
def mortality_index(row):
    if row['Mortality rate attributed to exposure to unsafe WASH services'] < 1:
        val = 14
    elif 1 <= row['Mortality rate attributed to exposure to unsafe WASH services'] <= 2:
        val = 13
    elif 2 < row['Mortality rate attributed to exposure to unsafe WASH services'] <= 3:
        val = 12
    elif 3 < row['Mortality rate attributed to exposure to unsafe WASH services'] <= 5:
        val = 11
    elif 5 <= row['Mortality rate attributed to exposure to unsafe WASH services'] <= 7.5:
        val = 10
    elif 7.5 <= row['Mortality rate attributed to exposure to unsafe WASH services'] <= 10:
        val = 9
    elif 10 < row['Mortality rate attributed to exposure to unsafe WASH services'] <= 20:
        val = 8
    elif 20 < row['Mortality rate attributed to exposure to unsafe WASH services'] <= 30:
        val = 7
    elif 30 <= row['Mortality rate attributed to exposure to unsafe WASH services'] <= 40:
        val = 6
    elif 40 < row['Mortality rate attributed to exposure to unsafe WASH services'] <= 50:
        val = 5
    elif 50 < row['Mortality rate attributed to exposure to unsafe WASH services'] <= 60:
        val = 4
    elif 60 <= row['Mortality rate attributed to exposure to unsafe WASH services'] <= 70:
        val = 3
    elif 70 < row['Mortality rate attributed to exposure to unsafe WASH services'] <= 80:
        val = 2
    elif 80 < row['Mortality rate attributed to exposure to unsafe WASH services'] <= 90:
        val = 1
    else:
        val = 0
    return val
```

```python
mortality_rate['mortality_index'] = mortality_rate.loc[mortality_rate['Granularity'] == 'Total'].apply(mortality_index, axis=1)
mortality_rate
```

```python
mortality_rate.to_csv("MortalityRateAttributedToWater.csv")
```

Accès aux services d'eau basique

```python
sns.histplot(x=water_services['Population using at least basic drinking-water services (%)'], bins=2000)
plt.xlim(99,100)
water_services.describe()
```

```python
def water_access_basic_index(row):
    if row['Population using at least basic drinking-water services (%)'] < 10:
        val = 0
    elif 10 <= row['Population using at least basic drinking-water services (%)'] <= 20:
        val = 1
    elif 20 < row['Population using at least basic drinking-water services (%)'] <= 30:
        val = 2
    elif 30 < row['Population using at least basic drinking-water services (%)'] <= 40:
        val = 3
    elif 40 <= row['Population using at least basic drinking-water services (%)'] <= 50:
        val = 4
    elif 50 < row['Population using at least basic drinking-water services (%)'] <= 60:
        val = 5
    elif 60 < row['Population using at least basic drinking-water services (%)'] <= 70:
        val = 6
    elif 70 <= row['Population using at least basic drinking-water services (%)'] <= 80:
        val = 7
    elif 80 < row['Population using at least basic drinking-water services (%)'] <= 90:
        val = 8
    elif 90 <= row['Population using at least basic drinking-water services (%)'] <= 92.5:
        val = 9
    elif 92.5 < row['Population using at least basic drinking-water services (%)'] <= 95:
        val = 10
    elif 95 < row['Population using at least basic drinking-water services (%)'] <= 97:
        val = 11
    elif 97 <= row['Population using at least basic drinking-water services (%)'] <= 98:
        val = 12
    elif 98 < row['Population using at least basic drinking-water services (%)'] <= 99:
        val = 13
    elif 99 < row['Population using at least basic drinking-water services (%)'] <= 100:
        val = 14
    else:
        val = None
    return val
```

```python
water_services['water_access_basic_index'] = water_services.apply(water_access_basic_index, axis=1)
water_services
```

```python
water_services.to_csv("BasicAndSafelyManagedDrinkingWaterServices.csv")
```
